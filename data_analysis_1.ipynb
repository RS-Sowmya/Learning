```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("List of Countries by Sugarcane Production.csv")
df.sample(5)
```

```python
#cleaning the data
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".", "")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".", "")
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".", "")
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].str.replace(".", "")
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(",", "")
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].str.replace(",", "")
```

```python
df.isna().sum()
```

```python
df[df["Acreage (Hectare)"].isna()]
```

```python
#drop null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.drop("Unnamed: 0", axis = 1, inplace = True)
df.head()
```

```python
#convert data types
df["Production (Tons)"] = df["Production (Tons)"].astype(float)
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].astype(float)
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].astype(float)
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.nunique()
```

```python
plt.subplot(2,2,1)
sns.boxplot(df["Production (Tons)"])
plt.subplot(2,2,2)
sns.boxplot(df["Production per Person (Kg)"])
plt.subplot(2,2,3)
sns.boxplot(df["Acreage (Hectare)"])
plt.subplot(2,2,4)
sns.boxplot(df["Yield (Kg / Hectare)"])
```

```python
plt.subplot(2,2,1)
sns.distplot(df["Production (Tons)"])
plt.subplot(2,2,2)
sns.distplot(df["Production per Person (Kg)"])
plt.subplot(2,2,3)
sns.distplot(df["Acreage (Hectare)"])
plt.subplot(2,2,4)
sns.plot(df["Yield (Kg / Hectare)"])
```

```python
df.describe()
```

```python
#check production
df_check = df[["Country", "Production (Tons)"]].set_index("Country")
df_check["Production (Tons)_percentage"] = df_check["Production (Tons)"]*100/df_check["Production (Tons)"].sum()
df_check
```

```python
df_check["Production (Tons)_percentage"].plot(kind = "pie", autopct = "%.2f")
```

```python
#to know the country with highest lands
dfacr_sort = df.sort_values("Acreage (Hectare)", ascending = False)
ax = sns.barplot(data = dfacr_sort.head(15), x = "Country", y = "Acreage (Hectare)")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.show()
```

```python
#same to find the yield
dfyield_sort = df.sort_values("Yield (Kg / Hectare)", ascending = False)
ax = sns.barplot(data = dfyield_sort.head(15), x = "Country", y ="Yield (Kg / Hectare)")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
plt.show()
```

```python
df_Continent = df.groupby("Continent").sum()
df_Continent
```

```python
df_Continent["Production (Tons)"].sort_values(ascending = False).plot(kind = "bar")
```

```python
df_Continent["Acreage (Hectare)"].sort_values(ascending = False).plot(kind = "bar")
```

```python
df["Continent"].value_counts() #Africa has highest no of countries but still land and production is high in South America
```

```python
#grouping one coloumn based on other
df_Continent["no_of_countries"] = df.groupby("Continent").count()["Country"]
df_Continent
```

```python
#how countries in continents effecting production
Continent_names = df_Continent.index.to_list()
sns.lineplot(data = df_Continent, x = "no_of_countries",y="Production (Tons)")
plt.xticks(df_Continent["no_of_countries"], Continent_names, rotation = 90 )
plt.show()
```
